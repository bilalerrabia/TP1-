```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(42)
temperature=np.random.uniform(40,100,100)
vibration=np.random.uniform(0.5,5.0,100)
courant=np.random.uniform(1,10,100)
duree_de_vie=5000-40*temperature-300*vibration-50*courant+np.random.normal(0,100,100)
data=pd.DataFrame({
    'temperature':temperature,
    'Vibration':vibration,
    'Courant':courant,
    'Duree de vie':duree_de_vie

})
data.to_csv('donnees_moteur.csv',index=False)
print(data.head())

donnees = pd.read_csv('donnees_moteur.csv')
#aafichez les statistiques descriptives at la correlation entre les variables
#visualisez la relation entre la temperature et la duree_de _vie avec un nuage de points.
# Affichage des statistiques descriptives
print("\nStatistiques descriptives :")
print(donnees.describe())

# Affichage de la matrice de corrélation
print("\nMatrice de corrélation :")
print(donnees.corr())

# Visualisation de la relation entre la température et la durée de vie avec un nuage de points
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temperature', y='Duree de vie', data=donnees)
plt.title('Relation entre la température et la durée de vie')
plt.xlabel('temperature')
plt.ylabel('Duree de vie')
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x=data[['temperature']]
y=data[['Duree de vie']]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
#evaluez les performances avec l'erreur quadratique moyenne et le coefficient R2
#affichez le modele de regression
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Évaluation des performances
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"MSE : {mse:.2f}, R² : {r2:.2f}")

# Équation du modèle
coefficient = model.coef_[0][0]
intercept = model.intercept_[0]
print(f"Équation : Duree de vie = {coefficient:.2f} * temperature + {intercept:.2f}")
```
